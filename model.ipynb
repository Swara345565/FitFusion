```python
# List of columns to drop
columns_to_drop = [
    'Data.Household Weights.1st Household Weight Description', 
    'Data.Household Weights.2nd Household Weight Description',
    'Data.Household Weights.1st Household Weight',
    'Data.Household Weights.2nd Household Weight'
]

# Drop the columns
data = df.drop(columns=columns_to_drop)
```

## Getting Insights From Data

```python
data.describe(exclude='object')
```

```python
data.describe(include='object')
```

```python
from sklearn.cluster import KMeans
import numpy as np

# Select relevant features for clustering (excluding non-numeric columns)
features = data.select_dtypes(include=['float64', 'int64'])

# Choose the number of clusters
num_clusters = 5  # You can adjust this number based on your needs
kmeans = KMeans(n_clusters=num_clusters)

# Fit the model
kmeans.fit(features)

# Assign cluster labels back to the original dataframe
data['Cluster'] = kmeans.labels_
```
